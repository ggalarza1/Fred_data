```python
from fredapi import Fred
fred = Fred(api_key = 'INSERT API KEY HERE')
import matplotlib.pyplot as plt
```

```python
#fetching the name of the series
series_info = fred.get_series_info('IMPMX')
title = series_info['title']
#print(title)

#You must specify the series ID of the data set you are fetchng
data = fred.get_series('IMPMX')

#Graphing the data
plt.plot(data.index, data.values)
plt.xlabel('Year')
plt.ylabel('Value (Millions $)')
plt.title(title)
plt.grid(True)
plt.show()
```

```python
import pandas as pd

#fetching the name of the series
series_info = fred.get_series_info('IMPMX')
title = series_info['title']
#print(title)

#You must specify the series ID of the data set you are fetchng
data = fred.get_series('IMPMX')
data = data[data.index >= pd.to_datetime('2020-01-01')]

#Graphing the data
plt.plot(data.index, data.values)
plt.xlabel('Year')
plt.ylabel('Value (Millions $)')
plt.title(title)
plt.grid(True)
plt.show()
```
